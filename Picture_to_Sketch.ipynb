### TASK 4: Image to Pencil Sketch with Python:

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python. 

```python
# Importing opencv
import cv2 
```

## Reading the original image from the system


```python
original_image = cv2.imread('IRONMAN.jpg')
# original_image = cv2.imread('SHANG_CHI.jfif')
```

```python
cv2.imshow("Color_Image", original_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Converting the RGB image to Gray scale image

Coverting to gray scale using cvtColor from cv2 library. 'cv2.cvtColor()' method is used to convert an image from one color space to another. There are more than 150 color-space conversion methods available in OpenCV
https://www.geeksforgeeks.org/python-opencv-cv2-cvtcolor-method/

```python
greyscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("GreyScale_Image", greyscale_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Converting the Gray scale image to Negative image

To decrease the complexity of the image and make the handling easier we will be converting the image to a grayscale image with the help of the cvtColor function.In other words Inversion of input array elements.
https://www.geeksforgeeks.org/arithmetic-operations-on-images-using-opencv-set-2-bitwise-operations-on-binary-images/

```python
# negative_image = 1 - original_image
negative_image = cv2.bitwise_not(greyscale_image)
```

```python
cv2.imshow("Negative_Image", negative_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Converting the Negative image to Blur image

* A negative of the image can be obtained by "inverting" the grayscale value of every pixel. Blurs an image using a Gaussian filter.The function convolves the source image with the specified Gaussian kernel.
* When a image is inverted it helps in studying the image with more precision and detail.
* The more detailed study is important for the sketch generation as the sketches need to be accurate and good.

```python
inv_image = cv2.GaussianBlur(negative_image, (15,15), 0)
blur_image = 255 - inv_image
```

```python
cv2.imshow("Blur_Image", blur_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Dividing Gray and Blur image 

* Now that the whole image processing is done, we will be passing the previous outputs  to make the appropriate changes to the image.
* The sketch image is created by using divide function in opencv by dividing the grayscale image and the blur image with the scale value of 256

```python
sketch_image = cv2.divide(greyscale_image, blur_image, scale = 256.0)
```

## Saving the image

```python
cv2.imwrite("Pencil_Sketch.jpg", sketch_image)
```

## Visualizing the final result

```python
cv2.imshow("Pencil_Sketch.jpg",sketch_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
